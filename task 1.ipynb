```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Yogita/Desktop/elevate lab 2nd intern/mall/Shopping Mall Customer Segmentation Data .csv", encoding="UTF-8")
df.shape
```

```python
# Preview the data
print("Initial Data Info:")
print(df.info())
print(df.head())
```

```python
#Identify and handle missing value

print("Missing Values Count:")
print(df.isnull().sum())

```

```python
#Remove duplicate rows

before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print("Removed {before - after} duplicate rows.")
```

```python
#Standardize text values


for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
print(df.head())
```

```python
#Convert date columns to a consistent format (dd-mm-yyyy)


# Detect date-like columns
for col in df.columns:
    if 'date' in col.lower() or 'dt' in col.lower():
        df[col] = pd.to_datetime(df[col], errors='coerce').dt.strftime('%d-%m-%Y')
```

```python
#  Rename columns to clean, uniform names


df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
print(df.head())
```

```python
# 7️⃣ Check and fix data types
# ------------------------------------------------------------

# Example: ensure Age is int, Income is float, etc.
if 'age' in df.columns:
    df['age'] = df['age'].astype(int)

# Convert all numeric-like columns
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass
```

```python
# Final check

print("Cleaned Data Info:")
print(df.info())
print(df.head())
```

```python
#Save cleaned data


df.to_csv("Shopping_Mall_Customer_Segmentation_Cleaned.csv", index=False)
print("Cleaned data saved as 'Shopping_Mall_Customer_Segmentation_Cleaned.csv'")
```
